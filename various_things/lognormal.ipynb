# Understanding lognormal distribution (in ERT)

```python
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

## First, generate 10000 "lognormal" distributed values with params 5 and 1

```python
generator = np.random.default_rng(42)
numbers = generator.lognormal(20, 2, 10000)
```

```python
print(f"mean of the raw numbers: {statistics.mean(numbers)}")
print(f"stdev of the raw numbers: {statistics.stdev(numbers)}")
print(f"natural log of the mean of the raw numbers: {np.log(statistics.mean(numbers))}")
print(f"natural log of the stdev of the raw numbers: {np.log(statistics.stdev(numbers))}")
print("NOTE: taking the log of the mean/stdev of the raw numbers is NOT the same as the mean/stdev of the log-transformed raw numbers")
plt.hist(numbers, 100)
plt.show()
```

## Then take the natural logarithm of these data, we expect to see a normal distribution with mean 5 and std-dev 1

```python
log_e_of_raw = np.log(numbers)
print(f"mean of the natural log of the raw numbers: {statistics.mean(log_e_of_raw)}")
print(f"stdev of the natural log of the raw numbers: {statistics.stdev(log_e_of_raw)}")
plt.hist(log_e_of_raw, 100)
plt.show()
```

## Also, do log10 of the "raw" numbers, to illustrate what we sometimes do in ERT (which I believe is confusing)

```python
log10_of_raw = np.log10(numbers)
print(f"mean of the log10 of the raw numbers: {statistics.mean(log10_of_raw)}")
print(f"stdev of the log10 of the raw numbers: {statistics.stdev(log10_of_raw)}")

plt.hist(log10_of_raw)
plt.show
```
